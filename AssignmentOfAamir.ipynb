```python
import matplotlib.pyplot as plt 
# Heights of bars
height = [10, 24, 36, 40, 5]

# Positions on the x-axis
left = [0, 1, 2, 3, 4]

# Labels for bars
names = ['winter Storm','Flood','Severe storm','Tropical Cyclone','Earthquicke']

# Colors for bars (must match the number of bars)
colors = ['red', 'green', 'blue', 'orange', 'purple']

# Plotting the bar chart
plt.bar(left, height, width=0.8, color=colors)

# Adding x-ticks (positions and labels)
plt.xticks(left, names)

# Naming the axes and the title
plt.xlabel('Disasters')
plt.ylabel('Deaths')
plt.title('Bar Chart')

# Showing the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
  
# defining labels 
activities = ['Total loss Cost', 'Deaths','Budget_asset'] 
  
# portion covered by each label 
slices = [3, 7, 8] 
# color for each label 
colors = ['r', 'g', 'b']   
# plotting the pie chart 
plt.pie(slices, labels = activities, colors=colors,  
startangle=90, shadow = True, explode = (0.1, 0, 0), 
radius = 1.2, autopct = '%1.1f%%') 
# plotting legend 
plt.legend() 
# showing the plot 
plt.show()
```
